```python
from openai import OpenAI
import pandas as pd

api_key = "sk-"
client = OpenAI(api_key=api_key)
```

```python
# Function to generate a study plan prompt for a given student
def generate_study_plan_prompt(student_data):
    prompt = (
    f"You are tasked with creating a personalized study plan for {student_data['Name']}, a {student_data['Year of study']} student in {student_data['Field of study']}, based on the following information:\n\n"
    f"Subjects: {student_data['List of subjects']}\n"
    f"Preferred Learning Styles: {student_data['Preferred Learning Styles']}\n"
    f"Personal Objectives: {student_data['Personal Objectives']}\n"
    f"Challenges: {student_data['Challenges']}\n"
    f"Extracurricular Activities: {student_data['Extracurricular activities']}\n\n"
    "Please generate a detailed timetable that addresses academic requirements, learning styles, personal objectives, challenges, and extracurricular activities. "
    "Ensure the timetable is flexible and caters to {student_data['Name']}'s unique needs.\n\n"
    "Additionally, consider adding some suggestions to help {student_data['Name']} overcome challenges and achieve objectives"
    "The timetable should be in the following format:\n\n"
    "| Day | Time | Activity | Focus Subject(s) | Learning Style | Objective/Challenge |\n"
    "|---|---|---|---|---|---| \n"
    "| | | | | |  | | | | | \n"
    "| ... | ... | ... | ... | ... | ... | \n"

)

    return prompt
```

```python
def main():
    # Load student sample data from CSV file
    csv_file = '/content/Student Data Rows.csv'  # Replace with your actual CSV file path
    students_df = pd.read_csv(csv_file)

    # As each row represent each student so we traverse
    for index, student_data in students_df.iterrows():
        prompt = generate_study_plan_prompt(student_data)
        print(f"Generating study plan for {student_data['Name']}...")

        # Make LLM call for response
        response = client.chat.completions.create(
            model="gpt-4",
            messages=[
                {"role": "system", "content": prompt},
                {"role": "user", "content": "Generate a study plan."},
            ],
            temperature=0.8,
            top_p=1,
        )

        study_plan = response.choices[0].message.content
        print(f"Study Plan for {student_data['Name']}:\n{study_plan}\n")

if __name__ == "__main__":
    main()
```
